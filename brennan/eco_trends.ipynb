### Using logistic regression, testing the importance of the stand-alone S&P data before adding other features

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn import metrics
```

```python
def make_Xy_shift(dframe, target, period): #, allfeatures):    
    features = list(dframe.columns.values)
    Xy = dframe.copy()
    Xy = kill_nulls(Xy)
    tag = target + '-' + str(period)
    Xy[tag] = Xy[target].shift(-period)
    binTag = tag + '-bin'
    Xy[binTag] = np.where(Xy[tag] > Xy[target], 1, 0)
    df_X = Xy.ix[:-1][features]
    #if allfeatures:
    #    df_X = Xy.ix[:-1][list(dframe.columns.values)]
    #else:
    #    df_X = Xy.ix[:-1][target]
    df_y = Xy.ix[:-1][tag]
    df_yBin = Xy.ix[:-1][binTag]
    return df_X, df_y, df_yBin
```

### Generate and test derivatives as possible indicators

A fundamental aspect of spotting trends is being able to classify trends as positive or negative. One way to do this is by using a differential of rolling means. In finance, this calculation is reffered to as a "MACD" or moving average convergence/divergence. According to [Investopedia](http://www.investopedia.com/terms/m/macd.asp), a MACD is defined as: 
> "A trend-following momentum indicator that shows the relationship between two moving averages of prices."
> 
> ![alt text](http://i.investopedia.com/inv/dictionary/terms/macd1.gif "macd chart")

### Calculate a simple MACD with the long period set at 200 days and the short period set at 50 days

```python
def find_smd(d, args):
    d = d.copy()
    smd = pd.stats.moments.rolling_mean(d, args[0]) - pd.stats.moments.rolling_mean(d, args[1])
    return smd

def rename_columns(d, label):
    d = d.copy()
    col = list(d.columns.values)     
    tag = [i + label for i in col] #for i in tickers]
    names = dict(zip(list(d.columns.values), tag))
    d.rename(columns=names, inplace=True)
    return d
```

```python
#old
def make_df(d, fn, args, drop):
    dframe = d.copy()
    col = list(dframe.columns.values)
    for i in col:
        dt = dframe[i].copy()
        if drop:
            dt = dt.dropna()
        dt = fn(dt, args)
        dframe[i] = dt
    return dframe
```

```python
df_smd = find_smd(df, [200,50])
lines = ['snp_close', 'intRate', 'dollar', 'gold', 'oil']
plot_lines(df_smd, lines, "Simple moving average differential: 200 vs 50 day")
```

### Calculating the exponentially weighted MACD with a long period of 26 days and a short period of 12 days

```python
def find_emd(d, args):
    emd = pd.stats.moments.ewma(d , span=args[0], freq="D") - pd.stats.moments.ewma(d , span=args[1], freq="D")
    return emd
```

### Calculating Bollinger values with a 20 day rolling period

One way to measure volatility is to compare moving averages with the standard deviation of price changes. This method was popularized in the 80's by John Bollinger who subsequently copyrighted the term Bollinger Bands®. When a price goes above or below two standard deviations from the moving average price change, it is outside of an upper or lower band and considered overbought or oversold. According to [Investopedia](http://www.investopedia.com/terms/b/bollingerbands.asp), a Bollinger calculations are useful: 
> "Because standard deviation is a measure of volatility, Bollinger Bands® adjust themselves to the market conditions. When the markets become more volatile, the bands widen (move further away from the average), and during less volatile periods, the bands contract (move closer to the average). The tightening of the bands is often used by technical traders as an early indication that the volatility is about to increase sharply."
> 
> ![alt text](http://i.investopedia.com/inv/dictionary/terms/bollingerband.gif "bollinger chart")

```python
def find_bolval(d, span): #redo with ewma?
    ma = pd.stats.moments.rolling_mean(d, span)  
    ab = abs(d - ma)  
    sd = pd.stats.moments.rolling_std(ab, span)  
    bolval = (d - ma) / sd #bollinger value
    #bollinger bands
    #upper = ma + 2*sd  #if d>upper, overbought
    #lower = ma - 2*sd  #if d<lower, oversold
    #width = 4*sd       #band width is total volatility
    return bolval
```
